```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_excel('/content/iris.xlsx')

# Select relevant features
# Ensure these column names exist in your dataset
X = data[['sepal_length', 'sepal_width']]

# Create and fit the K-Means model
kmeans = KMeans(n_clusters=3, random_state=42)  # Added random_state for reproducibility
kmeans.fit(X)

# Get cluster centers and labels
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualize the clusters
plt.scatter(X['sepal_length'], X['sepal_width'], c=labels, cmap='rainbow', label='Data Points')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], color='black', marker='x', s=100, label='Cluster Centers')
plt.xlabel('Sepal Length')  # X-axis label
plt.ylabel('Sepal Width')   # Y-axis label
plt.title('K-Means Clustering')  # Plot title
plt.legend()  # Add legend to identify data points and cluster centers
plt.show()
```
